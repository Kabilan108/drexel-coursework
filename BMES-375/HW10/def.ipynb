## GC-content  https://en.wikipedia.org/wiki/GC-content
### The GC pair is bound by three hydrogen bonds, while AT  pairs are bound by two.
### Caculate the GC-content
GC content is usually calculated as a percentage value and sometimes called G+C ratio or GC-ratio. GC-content percentage is calculated as Count(G + C)/Count(A + T + G + C) * 100%.


```python
my_dna = "ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT" 
length = len(my_dna) 
C_count = my_dna.count('C') 
G_count = my_dna.count('G')
print("length: " + str(length)) 
print("C count: " + str(C_count)) 
print("G count: " + str(G_count))
print("GC content: " + str((G_count+C_count)/length))
```
