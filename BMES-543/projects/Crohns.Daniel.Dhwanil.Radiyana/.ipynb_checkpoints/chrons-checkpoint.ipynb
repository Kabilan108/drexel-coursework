### Significantly Different miRNA between Controls and Patients Ordered by both P-Value and Fold Change

```python
[t_stat, p_val] = scipy.stats.ttest_ind(deltaCT_h, deltaCT_p, axis=1)
new_df = pd.DataFrame({"Fold Change":deltadeltaCT,"P-Value":p_val})

# Returns all differentially expressed miRNAs (p-value <0.01) and their fold changes (>=1.5)
all_de = new_df[ (abs(new_df['Fold Change']) >= 0.585) & (new_df['P-Value'] <= 0.05) ] # got results for diff expressed but fold change is diff
all_de.sort_values(by='P-Value')
all_de
```
