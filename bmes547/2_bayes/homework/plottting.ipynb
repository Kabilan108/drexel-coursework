```python
# plot a normal distribution
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

mean = 3
std = 1.5

x = np.arange(-3, 10, 0.1)

# Function to plot
gaussianpdf = lambda x, mean, std: np.exp( - (x - mean)**2 / (2* std **2)) / ( std * np.sqrt(2 * np.pi) )

# PDF with manual calculation
# y = np.exp( - (x - mean)**2 / (2* std **2)) / ( std * np.sqrt(2 * np.pi) )
y = gaussianpdf(x, mean, std)
plt.plot(x, y, 'r-', lw=2, label='manual')

# PDF with scipy.stats.norm
y = norm.pdf(x, mean, std)
plt.plot(x, y, 'b-', lw=10, label='norm.pdf', alpha=0.2)

# Show mean
plt.axvline(x=mean, color='k', linestyle='--', label='mean')

# Show +/- 2 std (95%) and show the area
plt.axvline(x=mean-2*std, color='k', linestyle=':', label='95%')
plt.axvline(x=mean+2*std, color='k', linestyle=':')
plt.fill_between(x, y, where=(x>mean-2*std) & (x<mean+2*std), color='g', alpha=0.2)

plt.legend()
plt.grid()
```
