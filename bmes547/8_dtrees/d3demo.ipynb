```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
#See bmes.ahmet/README.TXT for setting up BMESAHMETDIR environment variable.
import sys,os; sys.path.append(os.environ['BMESAHMETDIR']); import bmes
```

```python
datafile = bmes.downloadurl('http://sacan.biomed.drexel.edu/lib/exe/fetch.php?rev=&media=course:ml:d3:d3demo:weightgaindata.tab','weightgaindata.tab');
import pandas, numpy
data = pandas.read_table(datafile,sep='\t');
data.head(5)
X = data.values[:,:-1];
T = data.values[:,-1];
names=data.columns[:-1].to_list()

T=T>=0
```

```python
from sklearn import tree
d3 = tree.DecisionTreeClassifier()
d3 = d3.fit(X, T)
tree.plot_tree(d3);
```

```python
print(tree.export_text(d3, feature_names=names))
```
