```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import math


from scipy.integrate import solve_ivp
def ode45(*args,**kargs):
	#if not 'method' in kargs: kargs['method']='RK45';
	sol=solve_ivp(*args,**kargs)
	return sol.t,sol.y
```

```python
# Phase Plot
xmesh, ymesh = np.meshgrid(np.linspace(-8,8,20), np.linspace(-4,4,20))
dx=ymesh
dy=-np.sin(xmesh)
lens=np.sqrt(dx**2+dy**2)
dyu = dy/lens;
dxu = dx/lens;
plt.quiver(xmesh,ymesh,dxu,dyu,color=[.5,.5,1])
plt.xlabel('angle');
plt.ylabel('angular velocity')



# Solution Curve
def dxy(t, xy):
	return [xy[1], -math.sin(xy[0])];

T,XY = ode45(dxy, [0, 20], [-2, 1])
plt.plot(XY[0,:], XY[1,:], color='red')
```
